```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/summer_olympics/athlete_events.csv')
df.head()
```

```python
df.drop(['Name', 'Age', 'Height', 'Weight', 'Team', 'Games', 'Season', 'City'], axis=1, inplace=True)
df.rename(columns={'Sex': 'Gender', 'NOC': 'Country'}, inplace=True)
df.head()
```

```python
df.dropna(subset=['Medal'], inplace=True)
df.head()
```

```python
df = pd.get_dummies(df, columns=['Gender'])
df_year = df.groupby(['Year', 'Sport', 'Event']).sum()
df_year.head()
```

```python
df_year.rename(columns={'Gender_M': 'Gender_Men', 'Gender_F': 'Gender_Women'}, inplace=True)
```

```python
df_year.iloc[df.index.get_level_values(['Sport']).str.contains('Skiing')]
```
