```python
def country_df(file, codes):
        df = pd.read_csv('../data/' + file + '.csv')
        country = df.loc[0, '#country+name']
        table = pd.pivot_table(df, values='#indicator+value+num', 
                                index=['#country+name', '#date+year'], 
                                columns='#indicator+code', 
                                aggfunc=np.sum)
        women_df = table[codes]
        women_years_df = women_df.loc[(country, [2014, 2016]),]
        women_years_df.reset_index(inplace=True)
        return women_years_df
```

```python
female_indicator_codes = [23906, 24106, 48706, 49006, 120606, 
                                    123306, 123506, 136906, 169706, 169806, 
                                    175106, 175706, 57806, 112606, 31706, 36806]
```

```python
usa_df = country_df('usa', female_indicator_codes)
germany_df = country_df('germany', female_indicator_codes)
germany_df.head()
```

```python
multi_df = usa_df.copy()
multi_df.append(germany_df, ignore_index=True)
```
